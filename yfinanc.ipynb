```python
import yfinance as yf
import pandas as pd
```

```python
SP500 = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")
print(len(SP500))
```

```python
SP500 = SP500[0]
SP500_df = pd.DataFrame(SP500)
SP500_df
```

```python
SP500_all_tickers = SP500_df.Symbol.tolist()
print(SP500_all_tickers)
```

```python
# clean data so it can be passed to yahoo finance
# some stocks come across with a "." that needs to be changed to a "-"
stocks = []

for stock_ticker in SP500_all_tickers:
    ticker = stock_ticker.replace(".","-")
    stocks.append(ticker)
```

```python
# define a function download_data()
# name of the stock (key) - stock values (2015-2021) as the values
start_date = '2021-07-29'
end_date = '2021-07-30'

# define a function download_data()
def download_data():
    # name of the stock (key) - stock values (2015-2021) as the values
    stock_data = {}

    for stock in stocks:
        ticker = yf.Ticker(stock)
        stock_data[stock] = ticker.history(start=start_date, end=end_date)['Close']
    return pd.DataFrame(stock_data)

download_data()
```
