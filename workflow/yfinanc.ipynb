```python
import yfinance as yf
import pandas as pd
```

```python
SP500 = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")
print(len(SP500))
```

```python
SP500 = SP500[0]
SP500_df = pd.DataFrame(SP500)
SP500_df
```

```python
SP500_all_tickers = SP500_df.Symbol.tolist()
print(SP500_all_tickers)
```

```python
# clean data so it can be passed to yahoo finance
# some stocks come across with a "." that needs to be changed to a "-"
stocks = []

for stock_ticker in SP500_all_tickers:
    ticker = stock_ticker.replace(".","-")
    stocks.append(ticker)
```

```python
# define a function download_data()
# name of the stock (key) - stock values (2015-2021) as the values
start_date = '2021-07-26'
end_date = '2021-07-30'

# define a function download_data()
def download_data():
    # name of the stock (key) - stock values (2015-2021) as the values
    stock_data = {}

    for stock in stocks:
        ticker = yf.Ticker(stock)
        stock_data[stock] = ticker.history(start=start_date, end=end_date)['Close']
    return pd.DataFrame(stock_data)

download_data()
```

```python
SP500_df_trimmed = SP500_df[['Symbol', 'GICS Sector']]
SP500_df_trimmed.set_index(SP500_df_trimmed['Symbol'], inplace=True)
SP500_df_trimmed.drop(columns=['Symbol'], inplace=True)
SP500_df_trimmed
```

```python
data = yf.download(
    #tickers list or string as well
    tickers = stocks,

    # use "period" instead of start/end
    # valid periods: 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max
    # (optional, default is "1mo")
    period = "5y",

    # fetch data by interval (including intraday if period < 60 days)
    # valid intervals: 1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo
    # (optional, default is '1d')
    interval = '1d',

    # adjust all OHLC automatically
    # (optional, default is False)
    auto_adjust = True,

    # download pre/post regular market hours data
    # (optional, default is False)
    prepost = True,

    # use threads for mass downloading? (True/False/Integre)
    # (optional, default is True)
    threads = True,

    # proxy URL scheme use use when downloading?
    # (optional, default is None)
    proxy = None
)['Close']

data
```
